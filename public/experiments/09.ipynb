```python
import numpy as np
import matplotlib.pyplot as plt

omega = np.linspace(-500, 500, 500)
X_jw = 200 / (10000 + omega**2)

plt.figure(figsize=(8, 5)) # Ustaw rozmiar wykresu
plt.plot(omega, X_jw) # Rysujemy X_jw w funkcji omega
plt.title('Transformata Fouriera $X(j\omega)$ sygnału $e^{-100|t|}$') # Tytuł z symbolami LaTeXa
plt.xlabel('Częstotliwość $\omega$ (radiany/s)') # Etykieta osi X
plt.ylabel('$X(j\omega)$') # Etykieta osi Y
plt.grid(True) # Włącz siatkę
plt.show() # Wyświetl wykres
```
