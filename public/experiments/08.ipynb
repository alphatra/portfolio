```python
import numpy as np
import matplotlib.pyplot as plt

omega = np.linspace(-np.pi, np.pi, 500)
H_ejw = 1 / (1 - 0.9 * np.exp(-1j * omega))
magnitude_H = np.abs(H_ejw)
phase_H = np.angle(H_ejw)

plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 1)
plt.plot(omega, magnitude_H)
plt.title('Magnitude of H(e^(jw))')
plt.xlabel('Frequency (w)')
plt.ylabel('Magnitude')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(omega, phase_H)
plt.title('Phase of H(e^(jw))')
plt.xlabel('Frequency (w)')
plt.ylabel('Phase (radians)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
